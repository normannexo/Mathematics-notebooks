# Probability distributions

```python
from scipy.stats import lognorm, expon
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

## Continuous distributions

### Exponential distribution

$
f(x) = \lambda e^{-\lambda x} \\
F(x) =  1 - e^{-\lambda x}
$


```python
x = np.linspace(expon.ppf(0.001), expon.ppf(0.999))
plt.plot(x, expon.pdf(x, scale=3))
plt.plot(x, expon.cdf(x, scale=3))
```

### Log-normal distribution

$
\log X \sim N(\mu, \sigma^2) \\
f(x) = \frac{1}{\sigma x \sqrt{2\pi}}\cdot \exp\left(-\frac{(\log x - \mu)^2}{2\sigma^2}\right)
$

```python
x = np.linspace(0, 5)
plt.plot(x, lognorm.pdf(x, 1, 0, 1))
plt.plot(x, lognorm.pdf(x, 0, 1, 2))
```

# Probability distributions

```python
from scipy.stats import lognorm, expon, wei
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

## Continuous distributions

### Exponential distribution

$
f(x) = \lambda e^{-\lambda x} \\
F(x) =  1 - e^{-\lambda x} \\
X\sim \text{Exp}(1) \implies \frac{1}{\lambda} X ~ \text{Exp}(\lambda)\\  
\mathbb{E} X = \int_{0}^\infty x \lambda e^{-\lambda x} dx = \frac{1}{\lambda} \\  
\mathbb{V}(X) = \mathbb{E}(X^2)-(\mathbb{E}X)^2 = \frac{2}{\lambda^2}-\frac{1}{\lambda^2} = \frac{1}{\lambda^2}
$


```python
scale = 2
x = np.linspace(expon.ppf(0.001, scale=scale), expon.ppf(0.999, scale=scale))
plt.plot(x, expon.pdf(x, scale=scale))
plt.plot(x, expon.cdf(x, scale=scale))
plt.hlines(1, x[0], x[-1], linestyles='dashed')
```

### Weibull distribution
$
f(x) = \alpha \lambda x^{\alpha - 1} \exp(-\lambda x^\alpha)
$

### Log-normal distribution

$
\log X \sim N(\mu, \sigma^2) \\
f(x) = \frac{1}{\sigma x \sqrt{2\pi}}\cdot \exp\left(-\frac{(\log x - \mu)^2}{2\sigma^2}\right)
$

```python
x = np.linspace(0, 5)
plt.plot(x, lognorm.pdf(x, 1, 0, 1))
plt.plot(x, lognorm.pdf(x, 0, 1, 2))
```

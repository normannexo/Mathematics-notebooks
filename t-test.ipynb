```python
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set()
```

```python
#create data
mu_a = 3
sigma_a=4
data_a = np.random.normal(loc=mu_a, scale=4, size=1000)
sns.distplot(data_a)
plt.plot(range(-10, 15), stats.norm.pdf(range(-10,15), loc=mu_a, scale=sigma_a))
```

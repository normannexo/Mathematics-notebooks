# Differential equations with Python

```python
import numpy as np 
from scipy.integrate import odeint
import matplotlib.pyplot as plt 
%matplotlib inline
```

### Example 1
$
\frac{dy(t)}{dt} = - k y(t)
$

```python
def model(y,t, k):
    dydt = -k * y
    return dydt
```

```python
y0 = 5
t = np.linspace(0, 20)
k= 0.4
y1 = odeint(model, y0, t, args=(k,))
k= 0.5
y2 = odeint(model, y0, t, args=(k,))
k= 0.2
y3 = odeint(model, y0, t, args=(k,))

plt.plot(t, y1)
plt.plot(t, y2)
plt.plot(t, y3)
plt.xlabel('time')
plt.ylabel('y(t)')
```
